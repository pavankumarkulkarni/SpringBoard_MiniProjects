```python
def display_board(board,p1,p2):
    '''
    display the board. position of players in X and O. in 3 by 3 matrix.
    board is a list of positions selected by players. choices by players are appended one after other in the list.
    '''
    blank_board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    for i, v in enumerate(board):
        blank_board[v-1] = p1 if i%2 == 0 else p2
    print('''
    {} | {} | {}
    ----------
    {} | {} | {}
    ---------
    {} | {} | {}
    '''.format(*blank_board))
    
    
def game_end(board):
    '''
    Determine if the game ended. Game ends when player wins by completing any column or row or diagonal.
    Game ends in draw if all cells are filled and no player completes column, row or diagonal.
    input in list of selections by players appended in sequence.
    '''
    end_flag = False
    #Winning rules
    rule_conditions = [ '123', '456', '789', '147', '258', '369', '159', '357']
    rule_names = ['Row_1', 'Row_2', 'Row_3', 'Col_1','Col_2','Col_3', 'diagonal_1','diagonal_2']
    player1 = []
    player2 = []
    for i,v in enumerate(board):
        if i%2 == 0:
            player1.append(v)
        else:
            player2.append(v)
            
    player1.sort()
    player2.sort()
    player1_string = ''.join(str(p) for i,p in enumerate(player1))
    player2_string = ''.join(str(p) for i,p in enumerate(player2))
    if player1_string in rule_conditions:
        print('player 1 wins by  {}'.format(rule_names[rule_conditions.index(player1_string)]))
        end_flag = True
    elif player2_string in rule_conditions:
        print('player 2 wins by  {}'.format(rule_names[rule_conditions.index(player2_string)]))
        end_flag = True
    elif len(board) == 9:
        print('Game ended in draw !!')
        end_flag = True
    else:
        print('Continue the game')
        
    return(end_flag)
```

```python
continue_game = True
print('''
Welcome to the game of tic tac toe.
2 players  to play the game (both sitting at the same computer).
Position of the cells on the board to select are as below

    1|2|3
    -----
    4|5|6
    -----
    7|8|9

Player will win if any column, row or diagonal is fully occupied by same player.
If all cells are occupied the game ends in draw.

''')
while(continue_game):
    print(' !!! Let games begin !!!' )
    player_1 = input('Enter player 1 symbol choice X or O ? ')
    player_1 = player_1.upper()
    player_2 = 'X' if player_1 == 'O' else 'O'
    selections = []
    player = 0
    while(True):
        while(True):
            choice = int(input('Enter player {} position '.format(player%2 + 1)))
            if choice in selections:
                print('Position already occupied. Enter different position ')
                continue

            break

        player += 1
        selections.append(choice)
        display_board(selections, player_1,player_2)
        if not(game_end(selections)):
            continue
        continue_game = input('Do you want to start new game (Y/N) ? ')
        if(continue_game.upper() == 'Y'):
            continue_game = True
        else:
            continue_game = False
            print('Thanks for playing. Have a good Day!!')


        break
```
